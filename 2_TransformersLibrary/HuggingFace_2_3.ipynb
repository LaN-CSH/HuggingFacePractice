```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
tokenizer("Using a Transformer network is simple.","Thanks to HuggingFace!","Isn't it cool?","Yes!")
```

```python
tokenizer.save_pretrained("saving_folder")
```

```python
sequence = "Using a Transformer network is simple. Thanks to HuggingFace!"
tokens = tokenizer.tokenize(sequence)
print(tokens)
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

```python
print(tokenizer.encode(sequence))
```

```python
tokenizer.decode(ids)
encoded = tokenizer.encode(sequence)
tokenizer.decode([101, 7993, 170, 13809, 23763, 2443, 1110, 3014, 119, 102, 5749, 1106, 20164, 10932, 2271, 7954, 106, 102])
```
